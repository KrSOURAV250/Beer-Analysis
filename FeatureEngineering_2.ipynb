```python
from sklearn.preprocessing import OneHotEncoder
from tqdm import tqdm
import pandas as pd
import numpy as np
```

```python
beer = pd.read_csv(r"beer_for_FE_2.csv")
beer.sample(5)
```

### The columns `["min", "hour", "mday", "sec",	"year",	"wday",	"mon", "isdst",	"yday"]` are Categorical Data

```python
time = beer[["min", "hour", "mday", "sec",	"year",
             "wday",	"mon", "isdst",	"yday"]].astype("category")
time.info()
```

```python
timeCols = time.columns
timeCols
```

##### One-Hot Encoding on `["min", "hour", "mday", "sec",	"year",	"wday",	"mon", "isdst",	"yday"]`

```python
for column in tqdm(timeCols):
    encoder = OneHotEncoder(sparse_output=False)
    timeEncoded = encoder.fit_transform(np.array(time[column]).reshape(-1, 1))
    cols = list()
    for i in range(timeEncoded.shape[1]):
        cols.append(f"{column}_{i}")
    time = pd.concat([time.drop(columns=[column]), pd.DataFrame(
        timeEncoded, columns=cols).drop(columns=[cols[0]])], axis=1)
time
```

```python
time.info()
```

##### Merging into Original DataFrame

```python
beer = pd.concat([beer, time], axis=1)
beer
```

## Extracting Data for EDA 3

```python
beer.to_csv(path_or_buf="beer_EDA_3.csv", index=False)
```
